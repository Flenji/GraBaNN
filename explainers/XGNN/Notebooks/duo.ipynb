```python
import importlib
import sys
sys.path.append("../")
sys.path.append('../../../')
import explainers.GNNBoundary.gnn_boundary.models as models
from torch_geometric.data import Data
from utility_functions import *
import torch
#sys.path.append("./explainers/XGNN")
#sys.path.append("./graph_generation")
#import HouseSet as HouseSet
from XGNNInterface  import XGNNInterface
import matplotlib.pyplot as plt
```

```python
model = models.GCNClassifier(node_features=3,
                            num_classes=7,
                            hidden_channels=16,
                            num_layers=4)
```

```python
import os
cwd = os.getcwd()
file_path = os.path.abspath(os.path.join(cwd, '../../../model/report_models/duo.pt'))
model.load_state_dict(torch.load(file_path))
```

```python
mw = lambda x,e : model_wrapper(x,e,model=model)
encoding_dict = one_hot_encoding(3)
cgd = lambda nx_graph: convertNxToData(nx_graph, encoding_dict)
```

```python
li = [(0,0),(1,2),(2,1),(3,0),(4,2),(5,0),(6,1)]

color_dict = {(0,0,1):"red",(0,1,0):"blue",(1,0,0):"black"}

reward_stepwise_li = [0.1,0.2,0.3,0.4]#[0.1,0.2,0.3,0.4]
alpha_roll_out_li = [1,2,5]

#best_prob = 0
#best_graph = None
#best_rew_st = 0
#best_alph = 0

for cl, starting_node in li:

    best_prob = 0
    best_graph = None
    best_rew_st = 0
    best_alph = 0
    
    for reward_stepwise in reward_stepwise_li:
        for alpha_roll_out in alpha_roll_out_li:
            explainer = XGNNInterface(8, 50, cl, 100, 3, model = mw,\
                                      convertNxToData = cgd, starting_node=starting_node,\
                                          checkpoint=False, reward_stepwise=reward_stepwise, roll_out_alpha = alpha_roll_out)
            graph, prob = explainer.train()
            if prob > best_prob:
                best_prob = prob
                best_graph = graph
                best_rew_st = reward_stepwise
                best_alph = alpha_roll_out
    printGraph(best_graph, filename=f"./images/duo_xgnn_cl{cl}_starting_node_{starting_node}",clear_figure=True,color_dict=color_dict)
    print(f"\n################")
    print(f"class {cl} starting node {starting_node}")
    print(f"best prob = {best_prob}")
    print(f"best rew step {best_rew_st}")
    print(f"best alpha {best_alph}")
            
```
